```python
import requests
```

```python
def get_game(user):
    url= f'https://lichess.org/api/games/user/{user}'
    resp= requests.get(url=url, params ={'max':20}, headers={'Accept':'application/x-ndjson'})
    return resp.text
```

```python
res=get_game('Ositico')
```

```python
len(res.split('\n'))
```

```python
import json
gameee=([len(json.loads(game)['moves']) for game in res.split('\n') if game])
```

```python
from matplotlib import pyplot as plt
plt.plot(gameee)
```
